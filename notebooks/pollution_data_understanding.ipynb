```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("/Users/kanishjn/Downloads/air_pollution_data.csv")
weather_data = pd.read_csv("/Applications/my_work/Pm_monitor/data/processed/weather_data.csv")
weather_data.head()

PROJECT_ROOT = Path().resolve().parent
```

```python
df.info()
df.columns
print(df["city"].unique())
print(df["city"].unique())
df.shape
df.isna().mean()*100
```

```python
cities = ["delhi", "mumbai", "bengaluru", "kolkata", "hyderabad"]

df["city"] = df["city"].str.strip().str.lower()

df = df[df["city"].isin(cities)].copy()

df["date"] = pd.to_datetime(df["date"], errors='coerce')

df = df.sort_values(['city', 'date'])

df.to_csv(f"{PROJECT_ROOT}/data/processed/df.csv", index=False)

df.info()
df.isna().mean() * 100
```

```python
weather_data["date"] = pd.to_datetime(weather_data["date"])
weather_data["city"] = weather_data["city"].str.lower()
```

```python
#merge pollution data and weather data
merged_data = pd.merge(df, weather_data, on=["date","city"])
merged_data.head()
```

```python
merged_data.drop(columns=["Unnamed: 0","no","nh3"], inplace=True, errors="ignore")
merged_data.info()
print(merged_data.isna().mean()*100)
df.shape
```

```python
merged_data.to_csv(f"{PROJECT_ROOT}/data/processed/merged.csv")
```
